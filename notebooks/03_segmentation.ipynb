# 03 — Marketing Segmentation

Unsupervised segmentation built with `src/segment.py`. We cluster the population
using a marketing-actionable feature subset (life stage, employment intensity,
capital behaviour) and profile the resulting groups.

**Why not just split by the classifier's predicted income?**  The client asked
for segmentation for *marketing*, which is about offer/creative fit, not just
income. People with similar incomes can have very different life stages
(retiree vs young professional) and therefore very different marketing needs.

## Marketing read (see report for the full version)

* **High-earner segments** (weighted P(>\$50k) above ~25%): small population shares
  (~2–16%) but carry the purchase power — target with premium, financial, and
  aspirational creative.
* **Dual-income household spouses** and **young single urban workers** sit around
  the population mean — these are the volume play; run broad-reach campaigns
  here.
* **Retirees** (low positive rate, investment income, older): good candidates
  for health, travel, and value-oriented messaging.
* **Children** (two segments split by sex): not a direct marketing target for
  the retailer, but valuable as a household indicator if joining to parent
  records.
